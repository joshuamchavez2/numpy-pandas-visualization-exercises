```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(0, 10))
print(x)
```

```python
x = pd.Series(x).astype("float")
print(x)
```

```python
def equation(input_value):
    return (input_value**2)-input_value+2
    
```

```python
y = x.apply(equation)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.annotate('(0,0))', xy=(0, 0), xytext=(2, 15),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
y = x.apply(lambda x: math.sqrt(x))
```

```python
plt.plot(x,y)
plt.show()
```

```python
y = x.apply(lambda x: x**3)
plt.plot(x,y)
plt.show()
```

```python
y = x.apply(lambda x: 2**x)
plt.plot(x,y)
plt.show()
```

```python
y = x.apply(lambda x: 1/(x+1))
plt.plot(x,y)
plt.show()
```
