```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(0, 10))
print(x)
```

```python
x = pd.Series(x).astype("float")
print(x)
```

```python
y  = x.apply(lambda x: x**2 -x +2)
y
```

```python
def equation(input_value):
    return (input_value**2)-input_value+2
```

```python
y = x.apply(equation)
print(y)
```

```python
plt.plot(x,y)
plt.title('y=(x**2)-x+2')
plt.show()
```

```python
plt.plot(x,y)
plt.annotate('(0,0))', xy=(0, 0), xytext=(2, 15),
             arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
y = x.apply(lambda x: math.sqrt(x))
```

```python
plt.plot(x,y)
plt.title('y=x^.5')
plt.show()
```

```python
y = x.apply(lambda x: x**3)
plt.plot(x,y)
plt.title('y=x**3')
plt.show()
```

```python
y = x.apply(lambda x: 2**x)
plt.plot(x,y)
plt.title('y=2**x')
plt.show()
```

```python
y = x.apply(lambda x: 1/(x+1))
plt.plot(x,y)
plt.title('y=1/(x+1)')
plt.show()
```

```python
n_rows=2
n_cols=2


s1 = plt.subplot(n_rows, n_cols, 1)
s1.title.set_text('y=x^.5')
y = x.apply(lambda x: math.sqrt(x))
plt.plot(x, y)



s2 = plt.subplot(n_rows, n_cols, 2)
s2.title.set_text('y=x**3')
y = x.apply(lambda x: x**3)
plt.plot(x, y)



s3 = plt.subplot(n_rows, n_cols, 3)
s3.title.set_text('y=2**x')
y = x.apply(lambda x: 2**x)
plt.plot(x, y)


s4 = plt.subplot(n_rows, n_cols, 4)
s4.title.set_text('y=1/(x+1)')
y = x.apply(lambda x: 1/(x+1))
plt.plot(x, y)

plt.suptitle('Subplots')
plt.tight_layout()
```

```python
# Combine the figures you created in the last step into one figure where each of
#the 4 equations has a different color for the points.
#Be sure to include a legend and an appropriate title for the figure.
```

```python
x = list(range(0, 4))
x = pd.Series(x)

y = x.apply(lambda x: math.sqrt(x))
plt.plot(x, y, label="y=x^.5")

y = x.apply(lambda x: x**3)
plt.plot(x, y, label="y=x**3")

y = x.apply(lambda x: 2**x)
plt.plot(x, y, label="y=2**x")

y = x.apply(lambda x: 1/(x+1))
plt.plot(x, y, label="y=1/(x+1)")

plt.title('Equations')
plt.legend()
```
