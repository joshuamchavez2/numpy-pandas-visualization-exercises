```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(1, 3))
x = pd.Series(x).astype("float")

y = x.apply(lambda x : (0*x)+1)
plt.text(1.4, 1.1, 'y=0x+1', fontsize=10, color='blue')
plt.plot(x,y, c="blue")

y = x.apply(lambda x : math.log(x))
plt.text(1.5, 0.2, 'y=log(x)', fontsize=10, color='green')
plt.plot(x,y, c="green")

y = x.apply(lambda x : x)
plt.text(1.3, 2.2, 'y=x', fontsize=10, color='red')
plt.plot(x,y, c='red')

y = x.apply(lambda x : x * math.log(x))
plt.text(1.2, .5, 'y=x*log(x)', fontsize=10, color='black')
plt.plot(x,y, c="black")

y = x.apply(lambda x : x**2)
plt.text(1.1, 3.5, 'x**2', fontsize=10, color='cyan')
plt.plot(x,y, c="cyan")

y = x.apply(lambda x : 2**x)
plt.text(1.2, 2.7, '2**x', fontsize=10, color='magenta')
plt.plot(x,y, c = "magenta")

y = x.apply(lambda x : math.factorial(x))
plt.text(1.5, 1.6, 'x!', fontsize=13, color='saddlebrown')
plt.plot(x,y, c="saddlebrown")

y = x.apply(lambda x : x**x)
plt.text(1, 3.5, 'x**x', fontsize=10, color='slategrey')
plt.plot(x,y, c="slategrey")
            
plt.title("Big O Notation")
plt.xlabel('Elements')
plt.ylabel('Operations')
```
